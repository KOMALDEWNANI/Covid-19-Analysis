```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/WidhyaOrg/datasets/master/covid19.csv')
```

```python
df
```

```python
df1=df.groupby(['Date'],sort=False).sum()
```

```python
df1=df1.drop(['Sno'],axis=1)
```

```python
df1
```

```python
df1['Total Cases']=df1['ConfirmedIndianNational']+df1['ConfirmedForeignNational']+df1['Cured']+df1['Deaths']
```

```python
df1
```

```python
df1.reset_index(inplace=True)
```

```python
df1
```

```python
plt.plot(df1['Date'],df1['Total Cases'])
plt.xlabel('Date ')
plt.ylabel('Total Cases')
plt.title('Date vs Total Cases')
plt.show()
```

```python
#To calculate rate of increase
sum=0
for i in range(34,52):
    k=(df1.iloc[i-1,5]-df1.iloc[i,5])/df1.iloc[i,5]
    sum=sum+k
r=sum/18
print(r)
    
```

```python
#Function to predict forecast
def predict(p_o,r,t):
    p_t=p_o+np.exp(r*t) # p_o is the initial value and t is time period in days
    return p_t

```

```python
# To find p_t when p_0=31,t=26
p_t=predict(31,r,26)
p_t
```
